# Data Cleaning
Handleing Missing Values

Identifying columns with missing data.

Decides whether to drop rows/columns or impute values (mean, median, mode, or forward fill).


```python
df = df.drop_duplicates()
```

```python
 for col in df.select_dtypes(include='object').columns:
     df[col] = df[col].str.strip().str.lower()
```

```python
df.isnull().sum()
```

### droping the duplicate values from the data

```python
# Remove duplicates if any
if df.duplicated().sum() > 0:
    df = df.drop_duplicates()
    print("Duplicates removed")
else:
    # Drop rows with missing values if no duplicates
    df = df.dropna()
    print("Missing values dropped")
```

# def feature_engineering(df):
Step 4 of EDA: Feature Engineering & Transformation Creates new features, encodes categories, and scales data.

```python
# Ensure Order Date is datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Extract date parts
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day

print(df[['Order Date','Year','Month','Day']].head())
```

```python
# Profit Margin
if 'Sales' in df.columns and 'Profit' in df.columns:
     df['Profit Margin'] = df['Profit'] / df['Sales']
```

```python
# One-hot encode categorical columns
categorical_cols = df.select_dtypes(include='object').columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
print(" Feature Engineering Completed!\n")
print(" New Columns Added:", [col for col in df.columns if col not in ['Order Date','Sales','Profit','Quantity','Unit Price']])
```

```python
# Group by Year and Month
monthly_sales = df.groupby(['Year','Month'])['Sales'].sum().reset_index()

# Create a proper datetime column for plotting
monthly_sales['Period'] = pd.to_datetime(monthly_sales[['Year','Month']].assign(DAY=1))

# Plot
plt.figure(figsize=(10,6))
plt.plot(monthly_sales['Period'], monthly_sales['Sales'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract weekday names
df['Weekday'] = df['Order Date'].dt.day_name()

# Plot average sales by weekday
plt.figure(figsize=(8,5))
sns.barplot(x='Weekday', y='Sales', data=df, estimator='sum', order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.title("Sales by Weekday")
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie Chart: Sales by Category
if 'Category' in df.columns:
    category_sales = df.groupby('Category')['Sales'].sum()

    plt.figure(figsize=(7,7))
    plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=140)
    plt.title("Sales Distribution by Category")
    plt.show()
```

```python
# Pie Chart: Sales by Year
if 'Year' in df.columns:
    yearly_sales = df.groupby('Year')['Sales'].sum()

    plt.figure(figsize=(7,7))
    plt.pie(yearly_sales, labels=yearly_sales.index, autopct='%1.1f%%', startangle=140)
    plt.title("Sales Distribution Over the Years")
    plt.show()
```

```python
yearly_profit = df.groupby('Year')['Profit'].sum()
plt.pie(yearly_profit, labels=yearly_profit.index, autopct='%1.1f%%')
plt.title("Profit Distribution Over the Years")
plt.show()
```

```python
explode = [0.1 if year == df['Year'].max() else 0 for year in yearly_sales.index]
plt.pie(yearly_sales, labels=yearly_sales.index, autopct='%1.1f%%', explode=explode)
plt.title("Sales Distribution Over the Years (Highlight Latest Year)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=yearly_sales.index, y=yearly_sales.values, palette="viridis")
plt.title("Yearly Sales Comparison")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Group by Year
yearly_sales = df.groupby('Year')['Sales'].sum()

plt.figure(figsize=(8,5))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', linestyle='-')
plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```
